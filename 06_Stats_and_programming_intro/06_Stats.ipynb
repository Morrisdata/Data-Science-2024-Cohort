```python
'''
lets say you want to take any given point in your data and see how it fits with
all of the data or you want to know is my data wildly spread out or pretty uniform
let's look at how much or little it varys' ie measure the variability
to do this we will need to 
1) calculate the mean
2) calculate the deviation from the mean (n - mean)
3) calculate the abs of the average deviations    
3) calculate the mean of the deviations
'''
#lets keep it simple 
import numpy as np
import statistics as st
import math
x = [1,2,3,4,5]
x
```

```python
# calculate the mean
mean = st.mean(x)
mean
```

```python
# calculate the deviation from the mean( n - mean)
array = np.array(x)
x = array - mean
x = abs(x)
print(x)
```

```python
# calculate the mean of the deviations - variance
variance = st.mean(x)
print(variance)
```

```python
# square root of the avg. devation(variance) = standard devation
from math import sqrt
stddev = math.sqrt(variance)
print(stddev)
```

Standard deviation is the spread of a group of numbers from the mean. 


The variance measures the average degree to which each point differs from the mean. 
While standard deviation is the square root of the variance, variance is the average of all data points within a group

### Population vs Samples and Bessels Correction
We may not always be able to get all of the data and so we will rely on a sample
and then infer what the rest of the data will be. When this is done there is a risk that sample will skew towards normalcy and not represent the population of the dataset. Bessels correction accounts for this and is part of the variance calculation
* Bessels correction = Sum of the Square difference / (total entries -1)


```python
#pvariance is the variance of the entire population  
#large variance means the data is spread out smaller number 
#it is clustered closer
x = [1,2,3,4,5]
x= st.pvariance(x)
print("\n pvariance",x)
```

```python
# Population standard deviation pstdev() 
y = [1,2,3,4,5]
y = st.pstdev(y)
print(y)
```

```python
# sample variance
x = [1,2,3,4,5]
x = st.variance(x)
print(x)
```

```python
# sample and standard devaition
x = [1,2,3,4,5]
x = st.stdev(x)
print(x)
```

## Skew and Kurtosis

- A negative skewness value in the output indicates an asymmetry in the distribution corresponding to row 2 and the tail is larger towards the left hand side of the distribution.
- A positive skewness value in the output indicates an asymmetry in the distribution corresponding to row 3 and the tail is larger towards the right hand side of the distribution.

# PROBABILITY DENSITY FUNCTIONS (PDF and CDF(Cumulative))
'''
Relationship between an observation and the probability. So this is not just the probability but the relationship of the observation. 

Take for example a game and wanting to know the probability of winning.
PDF is saying you would win X number of times when playing the game. 
CDF is saying X or less games out of so many games. 
So if you had a 6 sided dice and wanted to pick a number lets say 1
you have a probability of winning .1 and a probability of losing .9

Now roll the dice 5 times. What are your chances of winning at least once?
How about what are your chances of winning 3 times or maybe 5 times.

What if we plotted this out on a histogram and found the chances of rolling the dice 5 times and not winning are 33% probability and then rolling the dice and winning once is about a 41% probability of winning and you end up with a distribution of probability that looks something like this
* Rolling 5 times and winning 0  33%
* Rolling 5 Times and winning 1  41%
* Rolling 5 Times and winning 2  20%
* Rolling 5 Times and winning 3  5%
* Rolling 5 times and winning 4  1%
* Rolling 5 times and winning 5  <1%

CDF asks well what would be the probability if I played X or less, lets say 2 or less times you then would add up the PDF outputs. So you would have a 94% chance of winning 2 or less times. 

This is the basic introductory concept
'''

```python
# Here is an example taken from simplilearn which has some nice resources to dive deep on this subject
# https://www.simplilearn.com/tutorials/statistics-tutorial/probability-density-function
#example of histogram using numpy random
# this is just and example for a random sample size using normal function
from matplotlib import pyplot
from numpy.random import normal
from numpy import mean
from numpy import std
from scipy.stats import norm
#generate sample
sample = normal(size= 1000)
#histogram
pyplot.hist(sample, bins = 10)
pyplot.show()
```

```python
# now lets assume that we don't know the output of what we just created. 
# so without know anthing about this we want to first calculate the mean and standard deviation
sample_mean = mean(sample)
sample_std = std(sample)
print('Mean={}, Standard Deviation={}'.format(sample_mean, sample_std))
```

```python
# these values can now be used to define the normal distro
dist = norm(sample_mean, sample_std)
dist
```

```python
# create sample probability for range of outcomes
values = [value for value in range(30, 70)]
probabilities = [dist.pdf(value) for value in values]
probabilities
```
