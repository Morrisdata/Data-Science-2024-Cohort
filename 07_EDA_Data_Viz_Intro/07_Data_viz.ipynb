
### Box Plot: show quartiles (and outliers) for one or more numerical variables

Five-number summary:

    min = minimum value
    25% = first quartile (Q1) = median of the lower half of the data
    50% = second quartile (Q2) = median of the data
    75% = third quartile (Q3) = median of the upper half of the data
    max = maximum value

(More useful than mean and standard deviation for describing skewed distributions)

Interquartile Range (IQR) = Q3 - Q1

Outliers:

    below Q1 - 1.5 * IQR
    above Q3 + 1.5 * IQR

#### Next steps
What has sparked your curiosity, what else is there to review in the data?
EDA is an iterative and ongoing process. As we continue through the course you will see various methods used to explore the data
we will begin grouping fields together, using boxplots to understand distribution. Can EDA be automated, yes, Should it be, no. EDA is for the benefit of the Analyst and is an exploration. There are templates you will create, modify, use and not use depending on your dataset and your questions. 
The general steps are:
- Describing your data both via metrics and annotation
- Observe and manipulate to explore data distributions
- create correlations or scatterplot matrix to understand relationship between variables. 
- Look for unusual patterns which could be outlier patterns or a repeated pattern
- Put your data into different groups and repeat the EDA process for the groups. 

Have tools for these steps, but take your time to walk the steps and understand each part. 


EDA works best when it generates deeper understanding and more questions.
